```python
from pyids.data_structures import (
    IDS,
    IDSRuleSet,
    ObjectiveFunctionParameters,
    IDSObjectiveFunction,
    mine_IDS_ruleset,
    mine_CARs
)

from pyarc.qcba.data_structures import QuantitativeDataFrame

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random

%matplotlib inline
```

```python
df = pd.read_csv("../data/titanic.csv")
```

```python
quant_df = QuantitativeDataFrame(df)
ids_rules = mine_IDS_ruleset(df, 1000)
```

```python
random.randint(1, 20)
```

```python
f_objectives = {
    "f0": [],
    "f1": [],
    "f2": [],
    "f3": [],
    "f4": [],
    "f5": [],
    "f6": [],
}

for _ in range(1000):
    print(_)
    
    sampling_max_n = len(ids_rules.ruleset)
    sampling_n = random.sample(1, sampling_max_n)
    ids_rules_sample = random.sample(ids_rules.ruleset, sampling_n)
    
    for rule in ids_rules_sample:
        rule.calculate_cover(quant_df)
    
    final_ruleset = IDSRuleSet(ids_rules_sample)
    
    params = ObjectiveFunctionParameters()
    params.params["all_rules"] = ids_rules
    params.params["len_all_rules"] = len(ids_rules)
    params.params["lambda_array"] = 7 * [1]
    params.params["quant_dataframe"] = quant_df
    
    obj_func = IDSObjectiveFunction(params)    
    
    f0 = obj_func.f0(final_ruleset)
    f1 = obj_func.f1(final_ruleset)
    f2 = obj_func.f2(final_ruleset)
    f3 = obj_func.f3(final_ruleset)
    f4 = obj_func.f4(final_ruleset)
    f5 = obj_func.f5(final_ruleset)
    f6 = obj_func.f6(final_ruleset)
    
    f_objectives["f0"].append(f0)
    f_objectives["f1"].append(f1)
    f_objectives["f2"].append(f2)
    f_objectives["f3"].append(f3)
    f_objectives["f4"].append(f4)
    f_objectives["f5"].append(f5)
    f_objectives["f6"].append(f6)

```

```python
f0_array = [f0, f0]
f1_array = [f1, f1]
f2_array = [f2, f2]
f3_array = [f3, f3]
f4_array = [f4, f4]
f5_array = [f5, f5]
f6_array = [f6, f6]
```

```python
N = 1
ind = np.arange(N) # the x locations for the groups
width = 0.2
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, f0, width)
ax.bar(ind, f1, width, bottom=f0)
ax.bar(ind, f2, width, bottom=f0)
ax.bar(ind, f3, width, bottom=f0)
ax.bar(ind, f4, width, bottom=f0)
ax.bar(ind, f5, width, bottom=f0)
ax.bar(ind, f6, width, bottom=f0)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0, 81, 10))
ax.legend(labels=['f0', 'f1', "f2", "f3", "f4", "f5", "f6"])
plt.show()
```

```python

plt.rcParams['font.size'] = 20

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))

recipe = ["Ostatn kritria", "$f_3$", "$f_6$", "$f_4$"]

data = [f0 + f1 + f4 + f6, f2, f5, f3]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

plt.show()
```
